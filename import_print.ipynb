```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

```python
h_town = np.load('/Users/arianauwaibi/Downloads/hs_houston.npy')

print('showing the size of image',h_town.shape)

type(h_town)

#band x pixels

# must be transformed into rows, columns, and pixels 

# Number of rows: 349  
# Number or Columns: 1905

# the final output should be (349,1905,144)
```

```python
#Transforming the image into a hypercube (rows, columns, and pixels ) 
h_town_cube = np.reshape(h_town.T,(349,1905,144))

#the T does the Transpose of the array so the image below comes out correctly

```

```python
print('showing the reshaped size of image',h_town_cube.shape)


plt.imshow(h_town_cube[:,:,10])
```

```python
_, fig = plt.subplots(nrows=4, ncols=1, figsize= (10,10))
for ax, image, channel in zip(fig,h_town_cube[:,:,(5,50,90,100)], [5,50,90,100]): 
    ax.set_axis_off()
    ax.imshow(image)
    ax.set_title("Channel: %i" % channel)

#cmap=plt.cm.gray_r
```
