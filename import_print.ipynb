```python
#Transforming the image into a hypercube (rows, columns, and pixels ) 
h_town_cube = np.reshape(h_town.T,(349,1905,144))

#the T does the Transpose of the array so the image below comes out correctly

```

```python
print('showing the reshaped size of image',h_town_cube.shape)


plt.imshow(h_town_cube[:,:,10])
```

```python
channels= (1,2,3,4)
_, fig = plt.subplots(nrows=4, ncols=1, figsize= (10,10))
for ax, image, channel in zip(fig,h_town_cube.T[:,:,:],channels):
    ax.set_axis_off()
    ax.imshow(image.T)
    ax.set_title("Channel: %i" % channel)

#dont forgert to Transpose the image 
```
